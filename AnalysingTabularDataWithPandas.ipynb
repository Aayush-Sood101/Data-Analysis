## Working With Dates

```python
covid_df.date # is an object which needs to be converted to datetime to be useful
```

```python
covid_df['date'] = pd.to_datetime(covid_df.date) # converting object to datetime
```

```python
covid_df['date']
```

```python
covid_df['year'] = pd.DatetimeIndex(covid_df.date).year
covid_df['month'] = pd.DatetimeIndex(covid_df.date).month
covid_df['day'] = pd.DatetimeIndex(covid_df.date).day
covid_df['weekday'] = pd.DatetimeIndex(covid_df.date).weekday
```

```python
covid_df
```

```python
covid_df_may = covid_df[covid_df.month == 5]

covid_df_may_metrics = covid_df_may[['new_cases' , 'new_deaths' , 'new_tests']]
```

## Grouping And Aggregation

```python
covid_month_df = covid_df.groupby('month')[['new_cases' , 'new_deaths' , 'new_tests']]
```

```python
covid_month_df # it is a new data frame
```

```python
covid_month_df.sum()
```

```python
covid_weekday_df = covid_df.groupby('weekday')[['new_cases' , 'new_deaths' , 'new_tests']]
```

```python
covid_weekday_df.sum() # aggregating by sum
```

```python
covid_weekday_df.mean() # can also be done by max min median etc
```

```python
covid_df['total_cases'] = covid_df.new_cases.cumsum()
covid_df['total_deaths'] = covid_df.new_deaths.cumsum()
covid_df['total_tests'] = covid_df.new_tests.cumsum()
```

```python
covid_df # cumulative value is NaN whenever Nan is encountered in org column
```

## Merging Data from multiple sources

```python
locations_df = pd.read_csv('C:\\Users\\AAYUSH SOOD\\Desktop\\Coding\\Data-Analysis\\location.csv')
```

```python
locations_df
```

```python
locations_df[locations_df.location == 'Italy'] # verification whether italy is present or not
```

```python
covid_df['location'] = 'Italy'
```

```python
covid_df
```

#### Requires at least one common column

```python
merged_df = covid_df.merge(locations_df , on = "location")
```

```python
covid_df
```

```python
merged_df['cases_per_million'] = merged_df.total_cases * 1e6 / merged_df.population
```

```python
merged_df['deaths_per_million'] = merged_df.total_deaths * 1e6 / merged_df.population
```

```python
merged_df['tests_per_million'] = merged_df.total_tests * 1e6 / merged_df.population
```

```python
merged_df
```

## Writing Data back to files

```python
result_df = merged_df[['date',
                      'new_cases',
                      'total_cases',
                      'new_deaths',
                      'total_deaths',
                      'new_tests',
                      'total_tests',
                      'cases_per_million',
                      'deaths_per_million',
                      'tests_per_million']]
```

```python
result_df
```
